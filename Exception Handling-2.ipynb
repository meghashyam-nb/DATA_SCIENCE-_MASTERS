Q3. What errors are defined in the ArithmeticError class? Explain any two with an example.

```python
#ZeroDivisionError: This exception is raised when division or modulo 
#operation is performed with zero as the divisor. It occurs when an attempt is made to divide a number by zero
numerator = 10
denominator = 0

try:
    result = numerator / denominator
    print("Result:", result)
except ZeroDivisionError:
    print("Error: Division by zero is not allowed.")

```

```python
#OverflowError: This exception is raised when an arithmetic calculation results in a value
#that exceeds the maximum representable value for a specific numeric type. 
#t typically occurs when performing calculations that lead to excessively large numbers.

import sys

try:
    value = sys.maxsize + 1
    print("Value:", value)
except OverflowError:
    print("Error: Arithmetic operation resulted in an overflow.")
```

Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

```python
#KeyError: This exception is raised when a dictionary key is not found during a lookup operation. 
#It occurs when an attempt is made to access a dictionary using a key that does not exist.
dictionary = {'a': 1, 'b': 2, 'c': 3}

try:
    value = dictionary['d']
    print("Value:", value)
except KeyError:
    print("Error: Key not found in the dictionary.")
```

```python
#IndexError: This exception is raised when a sequence (such as a list or a string) index is out of range.
#It occurs when an attempt is made to access an element from a sequence using an invalid index value

sequence = [1, 2, 3]

try:
    value = sequence[3]
    print("Value:", value)
except IndexError:
    print("Error: Index is out of range for the sequence.")
```

Q5. Explain ImportError. What is ModuleNotFoundError?

```python
#ImportError: This exception is raised when an error occurs during the import process. 
#It can occur in various situations, such as when a module or attribute is not found,
#when there are issues with module dependencies or circular imports, or when there are problems with the Python environment setup

try:
    import non_existent_module
except ImportError:
    print("Error: Module not found or cannot be imported.")
```

```python
#ModuleNotFoundError: Starting from Python 3.6, the ModuleNotFoundError exception is a subclass of ImportError. 
#It is specifically raised when a module or package cannot be found during the import process. 
#It provides a more precise error message to indicate that the specific module or package was not found.

try:
    import non_existent_module
except ModuleNotFoundError:
    print("Error: Module not found.")
```

Q6. List down some best practices for exception handling in python.

Best practices for exception handling in Python include catching specific exceptions, using multiple except blocks,
avoiding global exception handling, utilizing the finally block for cleanup, 
avoiding bare except statements, and providing meaningful error messages for debugging and troubleshooting.
